```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    y = x > 0
    return y.astype(np.int)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)
```

```python
if __name__ == '__main__': #Testing the functions
    x = np.arange(-10, 10, 0.01)
    y = step_function(x)
    plt.title('Step Function')
    plt.plot(x, y)
    plt.show()
    
    y = sigmoid(x)
    plt.title('Sigmoid Function')
    plt.plot(x, y)
    plt.show()
    
    y = relu(x)
    plt.title('ReLU')
    plt.plot(x, y)
    plt.show()
```
