```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def numerical_diff(f, x): #f: function, x : x where we want to get differential coefficient
    h = 10e-4
    return (f(x+h)-f(x-h)) / (2*h)
def diff(f, x):
    return numerical_diff(f, x)
def numerical_gradient(f, x):
    h = 1e-4
    grad = np.zeros_like(x)
    for i in range(len(x)):
        tmp = x[i]
        x[i] = tmp + h
        fx_plus_h = f(x)
        x[i] = tmp - h
        fx_minus_h = f(x)
        x[i] = tmp
        grad[i] = (fx_plus_h - fx_minus_h) / (2*h)
    return grad
```

```python
#Testing numerical diff function
if __name__ == '__main__':    
    x = np.arange(-100, 100, 1)
    f = lambda x: x**3
    y = f(x)
    plt.plot(x, y)
    plt.title('y = x^3')
    plt.show()
    
    y = diff(f, x)
    plt.plot(x, y)
    plt.title('y = (x^3)\'')
    plt.show()    
```
